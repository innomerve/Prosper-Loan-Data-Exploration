## Investigation Overview

> In this investigation, I wanted to look at the effects of borrower income range and  prosper score on the loan rate and original amount.


## Dataset Overview

> Here i analyze the prosper loans dataset. It contain 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and more. After the wrangling and removal of unused columns we have 113 937 rows and 14 columns.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
#function to convert to ordered categorical type
def convert_to_ordered_categorical_type(df, column, ordered_list):
    #convert columns into ordered categorical type
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                            categories = ordered_list)
    df[column] = df[column].astype(ordered_var)
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('cleanProsperLoanData.csv')



#convert categorials variables
#defining order
incomeRangeOrder = ["$0", "$1-24,999", "$25,000-49,999", "$50,000-74,999", "$75,000-99,999", "$100,000+"]
prosperScoreOrder= [number for number in range(1,11)]
#convert IncomeRange into ordered categorical type
convert_to_ordered_categorical_type(df,"IncomeRange",incomeRangeOrder)
#convert ProsperScore into ordered categorical type
convert_to_ordered_categorical_type(df,"ProsperScore",prosperScoreOrder)



#checking if it is correctly imported 
df.head(1)
```

```python
def barplot(x, y, xlabel, ylabel, title):
    """ ploting barplots """
    default_color = sb.color_palette()[0]
    sb.barplot(data=df, x=x, y=y, color=default_color)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=70)
```

## The borrower prosper score impact on the borrower rate
When searching the relationship of the prosper score and the borrower rate, we found that the worse is the posper score (10 to 1),  the higher is the borrower rate.


```python
plt.figure(figsize = [10, 6])
barplot('ProsperScore', 'BorrowerRate', xlabel="Prosper score", ylabel="Borrower rate", title="Average borrower rate by prosper score")
plt.tight_layout()
plt.show();
```

## The borrower prosper score impact on the loan original amount
The prosper score have also a relationship with the borrower loan amount. We found that the best is the prosper score (1 to 10), the higher is the loan original amount. So people with higner prosper score can borrow large amount of money.

```python
plt.figure(figsize = [10, 6])
barplot('ProsperScore', 'LoanOriginalAmount', xlabel="Prosper score", ylabel="Borrower loan amount ($)", title="Average loan original amount by prosper score")
plt.tight_layout()
plt.show();
```

## The borrower income range impact on the loan original amount

For the  income range and the loan original amount we found that the best is the income range, the higner is the loan original amount (except for $0 income range).

```python
plt.figure(figsize = [10, 6])
barplot('IncomeRange', 'LoanOriginalAmount', xlabel="Income range", ylabel="Borrower loan amount ($)", title="Average loan original amount by income range")
plt.tight_layout()
plt.show();
```

## The borrower income range impact on the borrower rate

For the  income range and the loan original amount we found that the best is the income range, the lower is the loan borrower rate (except for $0 income range).

```python
plt.figure(figsize = [10, 6])
barplot('IncomeRange', 'BorrowerRate', xlabel="Income range", ylabel="Borrower rate", title="Average loan borrower rate by income range")
plt.tight_layout()
plt.show();
```

## Impact of the borrower prosper score and income on loan rate and amount

At this level we noticed that more is the prosper score and the income range , the lower is the borrower rate and the higner is the loan original amount.

```python
g = sb.FacetGrid(data = df, col = 'ProsperScore', row='IncomeRange', size = 5)
g.map(plt.scatter, 'LoanOriginalAmount', 'BorrowerRate')
plt.suptitle('Impact of the borrower prosper score and income on loan rate and amount', size=30, y=1.12)
plt.tight_layout()
plt.show()
```
